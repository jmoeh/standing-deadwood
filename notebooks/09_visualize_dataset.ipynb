```python
register_df = pd.read_csv(register_file, index_col=0)
results_df = pd.read_csv(results_file)
```

```python
results_df["file_path"] = results_df["file_path"].str.strip()
```

```python
# join the two dataframes with suffixes for overlapping columns
merged_df = register_df.merge(results_df, on="file_path", how="left")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
# pivot table to get count of biome and resolution_bin
pivoted = merged_df.pivot_table(
    index="biome", columns="resolution_bin", values="f1", aggfunc="count"
)
sns.heatmap(pivoted, cmap="viridis", annot=True, fmt=".0f")
plt.show()
```

```python
# plot seaborn heatmap with biome and resolution as x and y and f1 as value
pivoted = merged_df.pivot_table(index="biome", columns="resolution_bin", values="f1")
sns.heatmap(pivoted, cmap="rocket", annot=True, fmt=".2f")
plt.show()
```
