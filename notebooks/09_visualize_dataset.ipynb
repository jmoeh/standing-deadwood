```python
import pandas as pd
import numpy as np
```

```python
register_file = "/net/scratch/jmoehring/tiles_register_biome.csv"
results_file = "/net/scratch/jmoehring/confusion_register.csv"
```

```python
register_df = pd.read_csv(register_file, index_col=0)
results_df = pd.read_csv(results_file)
```

```python
bins = np.arange(0.00, 0.22, 0.02)
register_df["resolution_bin"] = bins[np.digitize(register_df["resolution"], bins) - 1]
```

```python
register_df.to_csv("/net/scratch/jmoehring/tiles_register_biome_bin.csv")
```

```python
results_df["file_path"] = results_df["file_path"].str.strip()
```

```python
# join the two dataframes with suffixes for overlapping columns
merged_df = register_df.merge(results_df, on="file_path", how="left")
```

```python
merged_df.info()
```

```python
bins = np.arange(0.00, 0.22, 0.02)
merged_df["resolution_bin"] = bins[np.digitize(merged_df["resolution"], bins) - 1]
```

```python
# get mean of all not nan values f1
merged_df["f1"].dropna().mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
# plot seaborn heatmap with biome and resolution as x and y and f1 as value
pivoted = merged_df.pivot_table(index="biome", columns="resolution_bin", values="f1")
sns.heatmap(pivoted, cmap="mako")
plt.show()
```
